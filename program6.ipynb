Implement and demonstrate the FIND-S algorithm for finding the most specific hypothesis based on a given set of training data samples. Read the training data from a .CSV file.**

Objective: To implement the FIND-S algorithm and determine the most specific hypothesis that correctly classifies all positive training examples from a given dataset stored in a CSV file.

Step-by-Step Explanation

Load CSV dataset
Separate attributes and target.
Initialize the most specific hypothesis.
Process only positive examples.
Update hypothesis iteratively.
Display final hypothesis
Note: Creating CSV : Copy and Paste the below quoted Content in notepad and save it as .csv.

```python
import csv

data =list(csv.reader(open('train.csv')))
n = len(data[0])-1
h= ['0']*n
for row in data:
    if row[-1]=='yes':
        for i in range(n):
            if h[i]=='0':
                h[i]=row[i]
            elif h[i]!=row[i]:
                h[i]='?'
print("Hypothesis:",h)
```
