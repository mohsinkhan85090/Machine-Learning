```python
import nltk 
from nltk.corpus import brown
nltk.download('brown')
```

```python
categories = brown.categories()
```

```python
categories
```

```python
print(categories)
```

```python
import random

# Choose 5 random categories
random_categories = random.sample(list(categories), 5)

# Print words from each category
for category in random_categories:
    words = brown.words(categories=category)
    num_words = random.randint(10, 20)
    print(f"\nCategory: {category}")    
    print(" ".join(list(words)[:num_words]))
```

```python
# Count words in each category
word_counts = {}

for category in categories:
    words = brown.words(categories=category)
    word_counts[category] = len(words)

# Display the results
print("Word count for each category in Brown corpus:\n")
print(f"{'Category':<25} {'Word Count':>15}")
print("-" * 42)

for category, count in sorted(word_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"{category:<25} {count:>15,}")
```
