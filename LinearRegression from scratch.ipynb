Model Hypothesis


```python
def predict(x,m,b):
    return m*x + b
```

Cost Function(MSE)

```python
def compute_cost(y,y_hat):
    n = len(y)
    return (1/n)*sum((y-y_hat)**2 for y,y_hat in zip(y,y_hat))
 
```

Markdown

```python
def compute_gradient(x,y,y_hat):
    n = len(y)
    dm = (-2/n)*sum(x*(y-y_hat) for x,y,y_hat in zip(x,y,y_hat))
    db = (-2/n)*sum(y-y_hat for y,y_hat in zip(y,y_hat))
    return dm,db
```

update parameters

```python
def update_parameters(m,b,dm,db,alpha):
    m = m - alpha *dm
    b = b - alpha *db
    return m,b
```

model training

```python
def initialize_parameters():
    m = 0
    b = 0
    return m,b
def train(x,y, alpha=0.01, iterations=10):
    m,b = initialize_parameters()
    loss_history=[]
    for i in range(iterations):
        y_hat = predict(x,m,b) 
        loss=compute_cost(y,y_hat)
        loss_history.append(loss)
        dm,db = compute_gradient(x,y,y_hat)
        m,b = update_parameters(m,b,dm,db,alpha)
    return m,b, loss_history
```

Data Loading

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5],dtype=float)
y = np.array([7,11,15,19,23],dtype=float)
plt.scatter(x,y)
plt.xlabel('x')     
plt.ylabel('y')
plt.title('Scatter Plot of x vs y') 
plt.show()
```

Train model

```python
m , b ,loss_history = train(x,y,alpha=0.01,iterations=10)
print("\n final parameter:")
print("m:", m)
print("b:", b)  
plt.figure()
plt.plot(loss_history)
plt.xlabel('Iteration')     
plt.ylabel('Loss')
plt.title('Loss History')
plt.show()
```

```python
plt.figure()
y_pred= m*x + b
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')     
plt.show()
```
